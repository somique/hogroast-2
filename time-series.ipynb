## Setting up different similarity metrics

### Pearson correlation

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

overall_pearson_r = downsampled_dict["abituzumab"]['dipole']['Dipole_moment_all'].corr()
print(f"Pandas computed Pearson r:\n {overall_pearson_r}")
```

### Time-lagged cross-correlation

```python
def crosscorr(datax, datay, lag=0, wrap=False):
    """ Lag-N cross correlation.
    Shifted data filled with NaNs

    Parameters
    ----------
    lag : int, default 0
    datax, datay : pandas.Series objects of equal length
    Returns
    ----------
    crosscorr : float
    """
    if wrap:
        shiftedy = datay.shift(lag)
        shiftedy.iloc[:lag] = datay.iloc[-lag:].values
        return datax.corr(shiftedy)
    else:
        return datax.corr(datay.shift(lag))

d1 = downsampled_dict["abituzumab"]['dipole']['Dipole_moment_all']["300K"]
d2 = downsampled_dict["abituzumab"]['dipole']['Dipole_moment_all']["350K"]

max_lag = 50  # samples (not ps!)

rs = [crosscorr(d1, d2, lag) for lag in range(-max_lag, max_lag + 1)]

offset = max_lag - np.argmax(rs)

f, ax = plt.subplots(figsize=(14, 3))
ax.plot(range(-max_lag, max_lag + 1), rs)
ax.axvline(0, color='k', linestyle='--', label='Zero lag')
ax.axvline(-offset, color='r', linestyle='--', label='Peak synchrony')

ax.set(
    title=f'Offset = {offset} samples',
    xlabel='Lag (samples)',
    ylabel='Pearson r'
)
ax.legend()
```

One average, the signal at 300K is lagging behind the one at 350K

### Dynamic Time Warping

```python
import numpy as np
import matplotlib.pyplot as plt
from fastdtw import fastdtw
from scipy.spatial.distance import euclidean, cdist

# prepare series (ensure numpy arrays)
d1 = np.asarray(downsampled_dict["abituzumab"]['dipole']['Dipole_moment_all']["300K"])
d2 = np.asarray(downsampled_dict["abituzumab"]['dipole']['Dipole_moment_all']["350K"])

# If 1-D, reshape for cdist; if multi-D, keep shape as (n_features,)
if d1.ndim == 1:
    d1_for_cdist = d1.reshape(-1, 1)
else:
    d1_for_cdist = d1

if d2.ndim == 1:
    d2_for_cdist = d2.reshape(-1, 1)
else:
    d2_for_cdist = d2

# compute approximate DTW distance and path using fastdtw
d, path = fastdtw(d1, d2, dist=lambda a, b: abs(a - b))   # d is distance, path is list of (i,j) pairs

# compute full local cost matrix (Euclidean between points)
cost_matrix = cdist(d1_for_cdist, d2_for_cdist, metric='euclidean')

# compute accumulated cost matrix (standard dynamic programming)
n, m = cost_matrix.shape
acc_cost_matrix = np.full((n, m), np.inf)
acc_cost_matrix[0, 0] = cost_matrix[0, 0]

# first row/col
for i in range(1, n):
    acc_cost_matrix[i, 0] = cost_matrix[i, 0] + acc_cost_matrix[i-1, 0]
for j in range(1, m):
    acc_cost_matrix[0, j] = cost_matrix[0, j] + acc_cost_matrix[0, j-1]

# rest
for i in range(1, n):
    for j in range(1, m):
        acc_cost_matrix[i, j] = cost_matrix[i, j] + min(
            acc_cost_matrix[i-1, j],    # insertion
            acc_cost_matrix[i, j-1],    # deletion
            acc_cost_matrix[i-1, j-1]   # match
        )

# prepare path for plotting
path_x = [p[0] for p in path]
path_y = [p[1] for p in path]

plt.imshow(acc_cost_matrix.T, origin='lower', cmap='gray', interpolation='nearest')
plt.plot(path_x, path_y, 'w')   # white path line
plt.xlabel('temp1')
plt.ylabel('temp2')
plt.title(f'DTW Minimum Path (fastdtw) with distance: {np.round(d, 2)}')
plt.show()
```
